**Tipos de datos**

```
import pandas as pd

df = pd.read_csv('diabetes.csv')
df.dtypes
```

**Media, Mediana, Desviación Estándar, Mínimo y Máximo**

```
from google.colab import data_table

stats = df.agg(['mean', 'median', 'std', 'min', 'max'])
data_table.DataTable(stats, include_index=True)
```

**Valores nulos**

```
valores_faltantes = df.isnull().sum()
tabla_faltantes = pd.DataFrame({
    'Valores Nulos': valores_faltantes,
})

data_table.DataTable(tabla_faltantes, include_index=True)
```

**Valores en 0**

```
valores_faltantes = (df==0).sum()
tabla_faltantes = pd.DataFrame({
    'Valores Nulos': valores_faltantes,
})

data_table.DataTable(tabla_faltantes, include_index=True)
```

**Histograma Glucosa**

```
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")


sns.displot(data=df, x='Glucose', kde=True, color='skyblue', aspect=1.5)

plt.title('Distribución de la Variable Glucosa')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

**Histograma Edad**

```


sns.set_theme(style="whitegrid")


sns.displot(data=df, x='Age', kde=True, color='red', aspect=1.5)

plt.title('Distribución de la Variable Edad')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

**Histograma BMI**

```


sns.set_theme(style="whitegrid")


sns.displot(data=df, x='BMI', kde=True, color='green', aspect=1.5)

plt.title('Distribución de la Variable BMI')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

**`Box Plot Insulina`**

```
plt.figure(figsize=(8, 5))

sns.boxplot(x=df['Insulin'], color='lightgreen')

plt.title('Detección de Outliers en variable Insulina')
plt.show()
```

*(Los circulos grises representan los outliers)*

**`Box Plot Presión Sanguínea`**

```
plt.figure(figsize=(8, 5))

sns.boxplot(x=df['BloodPressure'], color='lightblue')

plt.title('Detección de Outliers en variable Presión Sanguínea')
plt.show()
```

*(Los circulos grises representan los outliers)*

**Heat Map de correlación de las variables**

```
corr_matrix = df.select_dtypes(include=['float64', 'int64']).corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Matriz de Correlación de Variables')
plt.show()
```

**Scatter plot entre edad y cantidad de embarazos**

```
import matplotlib.ticker as ticker

var1= 'Pregnancies'
var2= 'Age'

plt.figure(figsize=(8, 6))
ax = sns.regplot(data=df, x=var1, y=var2,
            scatter_kws={'alpha':0.5, 'color':'teal'},
            line_kws={'color':'red'})

ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

plt.title(f'Relación más fuerte detectada: Cantidad de embarazos vs Edad')
plt.show()
```
